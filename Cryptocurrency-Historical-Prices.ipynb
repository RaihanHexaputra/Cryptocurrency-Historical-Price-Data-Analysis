```python
# Import Library yang dibutuhkan
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# Mengumpulkan semua file CSV
folder_path = "C:\\Data Raihan\\Data Analyst Portofolio\\2 Cryptocurrency Historical Prices\\Dataset\\Cryptocurrency Historical Prices"
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
```

```python
# Membaca dan menggabungkan file CSV
df_list = [pd.read_csv(os.path.join(folder_path, file)) for file in csv_files]
combined_df = pd.concat(df_list, ignore_index=True)
```

```python
# Menyimpan hasil gabungan
combined_df.to_csv('Cryptocurrency_Historical_Price.csv', index=False)
```

```python
# mengecek apakah ada nilai yang tidak valid di semua kolom yang bersangkutan (nilai negatif atau null)
dataset[(dataset[['High', 'Low', 'Open', 'Close', 'Volume', 'Marketcap']] < 0) | (dataset[['High', 'Low', 'Open', 'Close', 'Volume', 'Marketcap']].isnull())]
```

```python
# Feature Engineering

# Menambahkan kolom rata-rata harga
dataset['AveragePrice'] = (dataset['High'] + dataset['Low']) / 2

# Menambahkan kolom persentase perubahan
dataset['PercentageChange'] = ((dataset['Close'] - dataset['Open']) / dataset['Open']) * 100

# Menampilkan dataset dengan fitur baru
dataset
```

```python
# Mengonversi kolom 'Date' menjadi datetime jika belum dilakukan
dataset['Date'] = pd.to_datetime(dataset['Date'])

# Mengecek format datetime
print(dataset['Date'].head())
```

```python
# Mengekspor dataset ke file CSV
dataset.to_csv('Cryptocurrency_Historical_Price.csv', index=False)
```
